```python
import numpy as np,pandas as pd
from decimal import Decimal
cancer = pd.read_excel('./Supplementary data 1.xlsx',index_col='SUBJECT_ID')
display(cancer.head(5))
# cancer['AFP'] = cancer['AFP'].str.replace(r'[^\d.]','', regex=True).astype(float)
cancer['AFP'] = cancer['AFP'].astype(float)
cancer['CA125'] = cancer['CA125'].astype(float)

cancer['CA19-9'] = cancer['CA19-9'].astype(float)
# cancer['AFP'] = cancer['AFP'].astype(float)
# cancer['afp_clean'] = cancer['AFP'].str.replace(r'[^\d.]', '', regex=True)

cancer['CA125'].head(5)
```

```python
X=cancer.drop('TYPE',axis = 1)
y=cancer['TYPE']
```

```python
X.shape
```

```python
##Normalization ##Min max normalization
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
print(X_scale)
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test, y_train ,y_test =train_test_split(X_scale,y, test_size=0.20,random_state=0)
```

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense

model =Sequential()
model.add(Dense(32 , activation ='relu' ,input_dim=49))
model.add(Dense(32,activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer = 'adam' , loss= 'binary_crossentropy' , metrics =['accuracy'])
model.fit(X_train,y_train,batch_size=32,epochs =15)
##steps per epoch = 279 / 32 ~ 9 steps in each epoch
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_pred =np.where(y_pred >0.5 ,1,0)
y_pred =y_pred.astype(int)
import numpy as np
np.column_stack((y_pred,y_test))
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

```python
##Regularization Techniques - Dropout , Early stopping , L1 and L2 regualarization ##
from tensorflow.keras.layers import Dropout
model = Sequential()

# model.add(Dropout(0.2 , input_dim=49)) Dropout at input layer
# model.add(Dense(32,activation='relu'))

model.add(Dense(32, activation='relu',input_dim =49))
model.add(Dropout(0.2))
model.add(Dense(32,activation='relu'))
model.add(Dropout(0.2))
```

```python
##early stopping
from keras.callbacks import EarlyStopping
model =Sequential()
model.add(Dense(32 , activation ='relu' ,input_dim=49))
model.add(Dense(32,activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer = 'adam' , loss= 'binary_crossentropy' , metrics =['accuracy'])

early_stop= EarlyStopping(monitor ='val_loss', patience = 5 , restore_best_weights=True)
model.fit(X_train,y_train,batch_size=32,epochs =100,\
validation_split = 0.2 ,callback=[early_stop] ,verbose=1)

y_pred= model.predict(X_test)
y_pred =np.where(y_pred >0.5 ,1,0)
y_pred =y_pred.astype(int)


print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

```python
##L1 and L2 regularization ##
from tensorflow.keras import layers , models, regularizers
model =Sequential()
model.add(Dense(32,activation ='relu', kernel_regularizer= regularizers.L1(0.001),input_dim=49))
model.add(Dense(32,activation ='relu', kernel_regularizer= regularizers.L1(0.001)))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer = 'adam' , loss= 'binary_crossentropy' , metrics =['accuracy'])
model.fit(X_train,y_train,batch_size=32,epochs =15)
```
